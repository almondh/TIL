# 시간데이터 형식변경 
test_interpolation.csv, weather_knn-All.csv 시간데이터형식을 시간별에서 일별 월별로 변경

```python
# load package
import pandas as pd # 데이터 전처리
import numpy as np # 데이터 전처리
import itertools

from datetime import datetime, timedelta # 시간 데이터 처리
```

## test_interpolation

```python
test = pd.read_csv("test_interpolation.csv")

test['Time'] = pd.to_datetime(test['Time']) 
test = test.set_index('Time')
```

```python
test.shape
```

```python
test.head()
```

```python
test_daily = test.groupby([lambda x: x.strftime("%Y-%m-%d")]).sum()
test_daily.to_csv('test_daily.csv', index=False)
```

```python
test_monthly = test.groupby([lambda x: x.strftime("%Y-%m")]).sum()
test_monthly.to_csv('test_monthly.csv', index=False)
```
